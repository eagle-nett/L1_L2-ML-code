## Bài Làm
### Lý Tiến Đạt

```python
# 1. Cài đặt các thư viện cần thiết
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import classification_report, accuracy_score
```

```python
# 2. Đọc dữ liệu từ file CSV
my_data = pd.read_csv("data/drug200.csv", delimiter=",")
```

```python
# 3. Tiền xử lý dữ liệu
# Lấy các thuộc tính từ dữ liệu
X = my_data[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
y = my_data['Drug']
```

```python
# 3.1 Chuyển đổi các trường categorical thành số
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F', 'M'])
X[:, 1] = le_sex.transform(X[:, 1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit(['LOW', 'NORMAL', 'HIGH'])
X[:, 2] = le_BP.transform(X[:, 2])

le_Chol = preprocessing.LabelEncoder()
le_Chol.fit(['NORMAL', 'HIGH'])
X[:, 3] = le_Chol.transform(X[:, 3])
```

```python
# 3.2 Chuẩn hóa thuộc tính Na_to_K
mean = X[:, -1].mean()
std = X[:, -1].std()
X[:, -1] = (X[:, -1] - mean) / std
```

```python
# 3.3 Chuyển đổi mục tiêu (y) thành số
le_drug = preprocessing.LabelEncoder()
le_drug.fit(['drugA', 'drugB', 'drugC', 'drugX', 'drugY'])
y = le_drug.transform(y)
```

```python
# 4. Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 5. Xây dựng mô hình Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
# 6. Đánh giá mô hình
y_pred = clf.predict(X_test)
print(f"Độ chính xác: {accuracy_score(y_test, y_pred)}")
print("Báo cáo Phân loại:")
print(classification_report(y_test, y_pred))
```

```python
# 7. Dự đoán cho bệnh nhân mới (ví dụ)
new_patient = np.array([['50', 'M', 'HIGH', 'HIGH', '25']])  # ví dụ bệnh nhân mới
new_patient[0][1] = le_sex.transform([new_patient[0][1]])[0]  # Chuyển đổi 'Sex' thành số
new_patient[0][2] = le_BP.transform([new_patient[0][2]])[0]  # Chuyển đổi 'BP' thành số
new_patient[0][3] = le_Chol.transform([new_patient[0][3]])[0]  # Chuyển đổi 'Cholesterol' thành số
new_patient[0][-1] = (float(new_patient[0][-1]) - mean) / std  # Chuẩn hóa 'Na_to_K'
```

```python
# Dự đoán loại thuốc cho bệnh nhân mới
prediction = clf.predict(new_patient)
predicted_drug = le_drug.inverse_transform(prediction)
print(f"Thuốc dự đoán cho bệnh nhân mới: {predicted_drug[0]}")
```

```python
# 1. Cài đặt các thư viện cần thiết
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import classification_report, accuracy_score

# 2. Đọc dữ liệu từ file CSV
my_data = pd.read_csv("data/drug200.csv", delimiter=",")

# 3. Tiền xử lý dữ liệu
# Lấy các thuộc tính từ dữ liệu
X = my_data[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
y = my_data['Drug']

# 3.1 Chuyển đổi các trường categorical thành số
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F', 'M'])
X[:, 1] = le_sex.transform(X[:, 1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit(['LOW', 'NORMAL', 'HIGH'])
X[:, 2] = le_BP.transform(X[:, 2])

le_Chol = preprocessing.LabelEncoder()
le_Chol.fit(['NORMAL', 'HIGH'])
X[:, 3] = le_Chol.transform(X[:, 3])

# 3.2 Chuẩn hóa thuộc tính Na_to_K
mean = X[:, -1].mean()
std = X[:, -1].std()
X[:, -1] = (X[:, -1] - mean) / std

# 3.3 Chuyển đổi mục tiêu (y) thành số
le_drug = preprocessing.LabelEncoder()
le_drug.fit(['drugA', 'drugB', 'drugC', 'drugX', 'drugY'])
y = le_drug.transform(y)

# 4. Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 5. Xây dựng mô hình Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# 6. Đánh giá mô hình
y_pred = clf.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print("Classification Report:")
print(classification_report(y_test, y_pred))

# 7. Dự đoán cho bệnh nhân mới (ví dụ)
new_patient = np.array([['50', 'M', 'HIGH', 'HIGH', '25']])  # ví dụ bệnh nhân mới
new_patient[0][1] = le_sex.transform([new_patient[0][1]])[0]  # Chuyển đổi 'Sex' thành số
new_patient[0][2] = le_BP.transform([new_patient[0][2]])[0]  # Chuyển đổi 'BP' thành số
new_patient[0][3] = le_Chol.transform([new_patient[0][3]])[0]  # Chuyển đổi 'Cholesterol' thành số
new_patient[0][-1] = (float(new_patient[0][-1]) - mean) / std  # Chuẩn hóa 'Na_to_K'

# Dự đoán loại thuốc cho bệnh nhân mới
prediction = clf.predict(new_patient)
predicted_drug = le_drug.inverse_transform(prediction)
print(f"Predicted drug for the new patient: {predicted_drug[0]}")

```

Giả sử bạn là một người nghiên cứu về y học và đã thu thập dữ liệu của các bệnh nhân mắc một loại bệnh. Trong quá trình điều trị, mỗi bệnh nhân sẽ phản ứng với một trong 5 loại thuốc A, B, C, X và Y.
